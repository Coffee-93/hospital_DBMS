```python
import pandas as pd
import numpy as np
```

```python
staff_type = ['a','b','c']

[np.random.choice(staff_type) for x in range(5)]
```

```python
staff_type = ['Cleaning Staff', 'Cooking Staff', 'Imaging Technician', 'Medical Assistant', 'Security', 'IT', 'Other' ]
```

```python
fake_staff = [
              {
               'Name': fake.name(),
               'Birthdate': fake.date(),
               'Address': fake.address(),
               'Staff_Type': np.random.choice(staff_type)   
              }
              for x in range(50)
]

fake_staff
```

```python
df = pd.DataFrame(fake_staff)
df
```

```python
df.to_csv('data.csv', index=True)
```
